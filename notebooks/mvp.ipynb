```python
%load_ext autoreload
%autoreload 2

import pickle
import numpy as np
from src.models.calcium_recording import CalciumRecording
from src.segmenter import Segmenter
import os
```

```python
def save_results(masks, flows, path="results"):
    np.save(f"{path}_masks.npy", masks)
    with open(f"{path}_flows.pkl", "wb") as f:
        pickle.dump(flows, f)

def load_results(path="results"):
    masks = np.load(f"{path}_masks.npy", allow_pickle=True)
    with open(f"{path}_flows.pkl", "rb") as f:
        flows = pickle.load(f)
    return masks, flows
```

```python
recording: CalciumRecording = CalciumRecording(path="../data/6s.tif")

if os.path.exists("results_masks.npy"):
    masks, flows = load_results()
else:
    masks, flows = Segmenter.generate_mask(recording)
    save_results(masks, flows)


print(len(masks))
print(len(flows))
print(flows)

recording.visualize(masks, flows)
```
